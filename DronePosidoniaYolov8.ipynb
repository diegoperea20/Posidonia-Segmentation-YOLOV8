```python
from ultralytics import YOLO
import glob
import cv2
import os
from google.colab.patches import cv2_imshow
import numpy as np

# Cargar el modelo entrenado
model = YOLO('/content/drive/MyDrive/runs/segment/yolov8n_seg_custom_transfer/weights/best.pt')

# Directorio que contiene las imágenes para validar
image_dir = '/content/drive/MyDrive/pos-2-6/valid/images'

# Obtener todas las imágenes en el directorio
image_files = glob.glob(os.path.join(image_dir, '*.jpg')) + glob.glob(os.path.join(image_dir, '*.png'))

# Crear un directorio para guardar los resultados
results_dir = 'validation_results_with_instancesMascaras'
os.makedirs(results_dir, exist_ok=True)

# Procesar cada imagen
for img_path in image_files:
    # Realizar la predicción
    results = model(img_path)

    # Leer la imagen original
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Obtener las máscaras de segmentación
    masks = results[0].masks
    boxes = results[0].boxes

    # Dibujar las máscaras y añadir IDs
    if masks is not None:
        for i, (mask, box) in enumerate(zip(masks, boxes)):
            # Convertir la máscara a un formato de imagen
            mask_img = mask.data.cpu().numpy().squeeze()
            mask_img = (mask_img * 255).astype(np.uint8)

            # Encontrar los contornos de la máscara
            contours, _ = cv2.findContours(mask_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

            # Generar un color aleatorio para esta instancia
            color = tuple(np.random.randint(0, 255, 3).tolist())

            # Dibujar el contorno
            cv2.drawContours(img_rgb, contours, -1, color, 2)

            # Aplicar la máscara con el color de la instancia
            colored_mask = np.zeros_like(img_rgb)
            colored_mask[mask_img > 0] = color
            img_rgb = cv2.addWeighted(img_rgb, 1, colored_mask, 0.5, 0)

            # Calcular el centro del contorno
            M = cv2.moments(contours[0])
            if M["m00"] != 0:
                cX = int(M["m10"] / M["m00"])
                cY = int(M["m01"] / M["m00"])
            else:
                cX, cY = 0, 0

            # Obtener la clase detectada
            class_name = model.names[int(box.cls)]

            # Añadir la clase y el ID
            cv2.putText(img_rgb, f"Clase: {class_name}, ID:{i+1}", (cX, cY), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    # Guardar la imagen con las detecciones
    output_path = os.path.join(results_dir, os.path.basename(img_path))
    cv2.imwrite(output_path, cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR))

    # Mostrar la imagen usando cv2_imshow
    cv2_imshow(img_rgb)

    # Mostrar información sobre las detecciones
    print(f"Detecciones para {os.path.basename(img_path)}:")
    for r in results:
        print(f"Número de objetos detectados: {len(r.boxes)}")
        for i, (box, mask) in enumerate(zip(r.boxes, r.masks)):
            print(f"  ID:{i+1}, Clase: {model.names[int(box.cls)]}, Confianza: {box.conf.item():.2f}")

    print("\n")  # Agregar una línea en blanco entre imágenes

print(f"Resultados guardados en {results_dir}")
```
